```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('anemia_dataset.csv')
df_cluster = pd.read_csv('clusters.csv')
df_severity = pd.read_csv('severity.csv')
```

```python
import os
print(os.listdir())
```

```python
print(len(df), len(df_cluster), len(df_severity))
df['cluster'] = df_cluster['cluster']
df['severity'] = df_severity['severity']
```

```python
df['cluster']
```

```python
print(df['severity'])
```

```python
df.sample(12)
```

```python
cluster_summary = df.groupby('cluster')['severity'].mean()
print(cluster_summary)
```

```python
cluster_order = cluster_summary.sort_values().index
print(cluster_order)
```

```python
mapping = {
    cluster_order[0]: 'Normal',
    cluster_order[1]: 'Mild',
    cluster_order[2]: 'Severe'
}

df['category'] = df['cluster'].map(mapping)
```

```python
cluster_summary = df.groupby('cluster')['severity'].mean()
print(cluster_summary)
```

```python
ranges = df.groupby('category')['severity'].quantile([0.1, 0.9]).unstack()
ranges.columns = ['low','high']
print(ranges)
```
