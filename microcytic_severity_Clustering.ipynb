```python
features = ['MCV','MCH','MCHC','RBC','HGB']
df = df[features]
```

```python
Q1 = df[features].quantile(0.25)
Q3 = df[features].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

```

```python
df_clean = df.copy()

for col in features:
    
    df_clean[col] = np.where(
        df_clean[col] < lower[col], lower[col],
        np.where(df_clean[col] > upper[col], upper[col], df_clean[col])
    )
```

```python
features = ['MCV','MCH','MCHC','RBC','HGB']

fig, ax = plt.subplots(1,5, figsize=(20,5))

for i, feature in enumerate(features):
    sns.boxplot(df_clean[feature], ax=ax[i], fill=True)
    ax[i].set_title(feature)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = pd.DataFrame(
    scaler.fit_transform(df_clean),
    columns=df_clean.columns
)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
# wcss :within cluster sum of square also refer as interia

wcss=[]

for i in range(1,20):
    km=KMeans(n_clusters=i)
    km.fit_predict(df_clean)
    wcss.append(km.inertia_)
```

```python
wcss
```

### ELBOW CURVE

```python
plt.plot(range(1,20), wcss)

plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal Number of Clusters")

plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2,8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(k, score)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=3)
y_agg = agg.fit_predict(X_pca)
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3, random_state=42)
y_gmm = gmm.fit_predict(X_pca)
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.5, min_samples=5)
y_db = db.fit_predict(X_pca)
```

```python
from sklearn.metrics import silhouette_score

print(silhouette_score(X_pca, y_means))  # KMeans
print(silhouette_score(X_pca, y_agg))    # Agglomerative
print(silhouette_score(X_pca, y_gmm))    # GMM
```

```python
def cluster_summary(df, cluster_col):
    return df.groupby(cluster_col)[['HGB','MCV','MCH','MCHC','RBC']].mean()
```

```python
df_clean['kmeans_cluster'] = y_means
print("KMeans:\n", cluster_summary(df_clean, 'kmeans_cluster'))
```

```python
df_clean['agg_cluster'] = y_agg
print("\nAgglomerative:\n", cluster_summary(df_clean, 'agg_cluster'))
```

```python
df_clean['gmm_cluster'] = y_gmm
print("\nGMM:\n", cluster_summary(df_clean, 'gmm_cluster'))
```

```python
df_clean['db_cluster'] = y_db

# remove noise (-1)
print("\nDBSCAN:\n", cluster_summary(df_clean[df_clean['db_cluster'] != -1], 'db_cluster'))
```

## Interpretation:
### KMeans = best separation
### Agglomerative = close (good validation)
### GMM = weak separation (data not very Gaussian-separated)
